### Cleaning Columns and Changing Datatypes

```python
# Clean price columns
for col in ['discounted_price','actual_price']:
    df[col] = (
        df[col]
        .str.replace('₹','')  # remove symbol (₹ )
        .str.replace(',','')  # remove commas (,)
        .astype(float)
    )
```

```python
# Clean discount_percentage (% sign → float)
df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('%','')  # remove symbol (%)
    .astype(float)
)
```

```python
# Clean rating (convert to float, coerce errors)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
# 1 Remove commas
df['rating_count'] = df['rating_count'].str.replace(',','', regex=True)

# 2 Convert to numeric (int)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# 3 impute nan with median
median_count = df['rating_count'].median()
df['rating_count'] = df['rating_count'].fillna(median_count)
```

```python
# impute rating col with mean
mean_rating = df['rating'].mean()
df['rating'] = df['rating'].fillna(mean_rating)
```

#### Handling Outliers

```python
def cap_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])
    return df

for col in ['discounted_price', 'actual_price', 'rating_count']:
    df = cap_outliers(df, col)
```

```python
from scipy import stats

def cap_outliers_z(df, col, threshold=3):
    z_scores = np.abs(stats.zscore(df[col]))
    median = df[col].median()
    df[col] = np.where(z_scores > threshold, median, df[col])
    return df
```
